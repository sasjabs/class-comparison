```python
from Cross import Cross
```

Определение входных и выходных файлов, выполнение расчёта

```python
cr = Cross()
rasters = ['data/Class1.tif', 'data/Class2.tif']
outputRaster = 'output/CrossMap.tif'
outputTable = 'output/CrossTable.csv'
cr.Execute(rasters, outputRaster, outputTable)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
```

```python
data = cr.arrays[0]
values = np.unique(data)
values = values[values!=0]

plt.figure(figsize=(15,12))
im = plt.imshow(data, interpolation='None', cmap = 'tab10');

colors = [im.cmap(im.norm(value)) for value in values]
patches = [mpatches.Patch(color=colors[i], label=f'Class {values[i]}') for i in range(len(values))]
plt.legend(handles=patches, bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0)

plt.show()
```

```python
data = cr.arrays[1]
values = np.unique(data)
values = values[values!=0]

plt.figure(figsize=(15,12))
im = plt.imshow(data, interpolation='None', cmap = 'tab10');

colors = [im.cmap(im.norm(value)) for value in values]
patches = [mpatches.Patch(color=colors[i], label=f'Class {values[i]}') for i in range(len(values))]
plt.legend(handles=patches, bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0)

plt.show()
```

```python
data = cr.cross
values = np.unique(data)
values = values[values!=0]
reclass = cr.reclass

plt.figure(figsize=(15,12))
im = plt.imshow(data, interpolation='None', cmap='tab10');

colors = [im.cmap(im.norm(value)) for value in values]
patches = [mpatches.Patch(color=colors[i], label=f'Class {reclass[values[i]]}') for i in range(len(values))]
plt.legend(handles=patches, bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0)

plt.show()
```
